```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import random
import seaborn as sns
import os
```

```python
datos=pd.read_csv(os.getcwd()+"/titanic.csv", delimiter=",")
datos.head()
```

```python
datos.isnull().sum()
```

```python
Edad=datos["Age"]
Edad_no_nulos=datos["Age"].dropna()
promedio=Edad_no_nulos.mean()
# rellenar todos los valores vacios
datos["Age"].fillna(promedio, inplace=True)
datos.isnull().sum()
```

```python
Cabinete=datos["Cabin"]
lista_de_valores=datos["Cabin"].dropna().tolist()
for idx, valor in Cabinete.items():
    if pd.isna(valor):
        indice=random.randint(0, len(lista_de_valores)-1)
        datos.at[idx, "Cabin"]=lista_de_valores[indice]
datos.isnull().sum()
```

```python
Embarcacion=datos["Embarked"]
lista_de_valores=datos["Embarked"].dropna().tolist()
for idx, valor in Embarcacion.items():
    if pd.isna(valor):
        indice=random.randint(0, len(lista_de_valores)-1)
        datos.at[idx, "Embarked"]=lista_de_valores[indice]
datos.isnull().sum()
```

```python
datos.head()
```

```python
# Contar la cantidad de sobrevivientes y no sobrevivientes
sobrevivientes=datos['Survived'].value_counts()

# Crear el gráfico de barras
plt.bar(['No sobrevivió', 'Sobrevivió'], sobrevivientes)
plt.xlabel('Sobrevivientes')
plt.ylabel('Cantidad')
plt.title('Cantidad de pasajeros sobrevivientes vs. no sobrevivientes')
plt.show()
```

```python
# Contarla cantidad de personas por cada clase
clases=datos["Pclass"].value_counts()
# Utilizar esoso valores para crear una grafica de pie que represente esos datos
plt.pie(clases, labels=["Clase 1", "Clase 2", "Clase 3"])
plt.title('Cantidad de pasajeros de clase 1, 2 y 3')
plt.show()
```

```python
valores_edad=datos["Age"].to_list()
intervalos=int(np.round(np.sqrt(len(valores_edad))))
plt.hist(valores_edad, bins=intervalos, color="gray")
plt.xlabel("Edades de los pasajeros")
plt.ylabel("Frecuencia")
plt.title("Histograma de edades y frecuencias de los pasajeros")
plt.show()
```

```python
Generos=datos["Sex"].value_counts()
plt.pie(Generos, labels=["Hombres", "Mujeres"])
plt.title("Proporcion de hombres y mujeres en el barco")
plt.show()
```
