```python
import random
arreglo=[random.randint(0, 500) for _ in range(30)] # Generar 30 valores diferentes entre 0 a 500 y almacenarlas en una lista
numero_bins=6
n, bins, patches= plt.hist(arreglo, numero_bins, facecolor="gray")
plt.show()
```

```python
import numpy as np
n = 12 
X = np.arange(n) 
Y1 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n) 
Y2 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n) 
plt.bar(X, +Y1, facecolor='#9999ff', edgecolor='white') 
plt.bar(X, -Y2, facecolor='#ff9999', edgecolor='white') 
for x, y in zip(X, Y1): 
    plt.text(x + 0.4, y + 0.05, '%.2f' % y, ha='center', va='bottom') 
plt.ylim(-1.25, +1.25)
```

```python
labels = 'Bananas', 'Peras', 'Naranjas', 'Limones'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0.1) # vamos a separar del pastel al segundo y cuarto elemento
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',shadow=True, startangle=90)
#ax1.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C, S = np.cos(X), np.sin(X)
plt.plot(X, C, color="r", linestyle="-.")
plt.plot(X, S)
plt.show()
```
