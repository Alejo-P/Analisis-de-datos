```python
import numpy as np # Importar numpy para utilizar la funcion random
import math # Importar math para usar funciones matematicas
# Generar 30 valores diferentes entre 0 a 500 y almacenarlas en una lista
arreglo=[np.random.randint(0, 500) for _ in range(30)]
# Especificar el numero de intervalos (columnas) que dispondra la grafica
intervalos=int(round(math.sqrt(len(arreglo))))
# Graficar el histograma
n, bins, patches= plt.hist(arreglo, intervalos, facecolor="dodgerblue")
# mostrar el histograma
plt.show()
```

```python
# Indicar el numero de barras a graficar
n_barras = 12 
# Crear un arreglo unidimensional con valores desde 0 hasta 12-1
X = np.arange(n_barras)
# Generar dos conjuntos de datos Y1 y Y2 utilizando números aleatorios
Y1 = (1 - X / float(n_barras)) * np.random.uniform(0.5, 1.0, n_barras) 
Y2 = (1 - X / float(n_barras)) * np.random.uniform(0.5, 1.0, n_barras) 
# Graficar las barras correspondientes
plt.bar(X, +Y1, facecolor='#9999ff', edgecolor='white') # Barras positivas (en direccion hacia arriba y de color azul)
plt.bar(X, -Y2, facecolor='#ff9999', edgecolor='white') # Barras negativas (en direccion hacia abajo y de color rojo)
# Ubicar etiquetas de texto en la parte superior de cada barra
for x, y in zip(X, Y1): 
    plt.text(x, y + 0.05, '%.2f' % y, ha='center', va='bottom') 
# Ubicar etiquetas de texto en la parte inferior de cada barra
for x, y in zip(X, Y2):
    plt.text(x, -y - 0.05, '%.2f' % -y, ha='center', va='top')
# Establecer limites de valores para ej eje y
plt.ylim(-1.25, +1.25)
# Mostrar la grafica
plt.show()
```

```python
# Generar listas de 30 números enteros aleatorios en el rango de 0 a 25 para las coordenadas x e y
x=[np.random.randint(0, 25) for _ in range(30)]
y=[np.random.randint(0, 25) for _ in range(30)]
# Crear un gráfico de dispersión a partir de los puntos (coordenadas) generados
plt.scatter(x, y)
# Mostrar la grafica
plt.show()
```

```python
# Generar un arreglo unidimensional de 20 valores desde 0 hasta 1
Z=np.random.uniform(0, 1, 20)
# Utilizar el arreglo creado para graficar un diagrama de pie
plt.pie(Z)
# Mostrarla grafica
plt.show()
```

```python
# Definir un diccionario que contenga los elementos de la grafica (etiqueta y tamaño)
Elementos = {'Bananas':15,'Peras':30,'Naranjas':45, 'Limones':10 }
# Definir la lista "explode" para separar porciones específicas del pastel
explode = (0, 0.1, 0, 0.1) # Vamos a separar del pastel al segundo y cuarto elemento
# Crear una figura y ejes para el gráfico de pastel
fig1, ax1 = plt.subplots()
# Crear el gráfico de pastel utilizando el método pie
ax1.pie(list(Elementos.values()),  # Tamaños de las porciones
        explode=explode,  # Lista de explosiones
        labels=list(Elementos.keys()),  # Etiquetas de las porciones
        autopct='%1.1f%%',  # Formato de porcentaje en cada porción
        shadow=True,  # Agregar sombras
        startangle=90)  # Ángulo inicial de la primera porción
# Ajustar el aspecto para que el pastel se dibuje como un círculo
ax1.axis('equal')
plt.show()
```

```python
# Generar un conjunto de valores X desde -π hasta π con 256 puntos
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
# Calcular el seno y coseno de cada valor en X
C, S = np.cos(X), np.sin(X)
# Graficar los respectivos valores
plt.plot(X, C, color="r", linestyle="-.")
plt.plot(X, S)
# Mostrar la grafica
plt.show()
```
