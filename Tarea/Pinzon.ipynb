*Debido a que la columna "Cabin" no tiene datos numericos, se procedio a rellenar los valores nulos de otra manera*

```python
import random #Generar numeros aleatorios
Elementos_pd=set_csv['Cabin'].dropna() #Extraer valores no nulos de la columna
Elementos={} #Diccionario donde se almacenara el indice desde 0 hasta el numero de valores, con el valor de los valores extraidos
Cabina=set_csv['Cabin'] #Extraer del DataFrame la columna "cabin"
ind=0 #Inicializar el indice en 0
#Iterar entre clave y valor de los valores extraidos y almacenarlos en el diccionario
for clave, valor in Elementos_pd.items():
    Elementos[ind]=valor
    ind+=1

# Iterar sobre todos los valores de la columna extraida
for i, valor in Cabina.items():
    if pd.isna(valor): #Comprobar que el valor no sea nulo
        indice=random.randint(0, 90) #Generar un numero aleatorio entre 0 a 90
        set_csv.at[i, "Cabin"]=Elementos[indice] #Acceder a un valor aleatorio del diccionario y establecerlo al indice con valor nulo
#Comprobar existencia de valores nulos
set_csv.isnull().sum()
```
