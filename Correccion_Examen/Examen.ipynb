_Etapa 3 - Limpieza de datos_

```python
Combinacion=pd.read_csv(archivos+"/Datos.Combinacion.csv", sep=",")
Combinacion
```

```python
Combinacion.isnull().sum()
```

```python
Disponibilidad=Combinacion['Available Globally?']
for clave, valor in Disponibilidad.items():
    if pd.isna(valor):
        Combinacion.at[clave, "Available Globally?"]="Yes"
Combinacion.isnull().sum()
```

```python
Columna="Number of Ratings"
Disponibilidad=Combinacion[Columna]
for clave, valor in Disponibilidad.items():
    if pd.isna(valor):
        Combinacion.at[clave, Columna]="Yes"
Combinacion.isnull().sum()
```

```python
import random
Lanzamiento=Combinacion['Release Date']
for clave, valor in Lanzamiento.items():
    if pd.isna(valor):
        mes=random.randint(1, 12)
        if mes==2: # Comprobar los meses (ya que no todos tienen 30 dias)
            dia=random.randint(1, 28)
        else:
            dia=random.randint(1, 30)
        # Formatear la cadena final y agregarla a la columna y posicion del valor nulo
        Combinacion.at[clave, "Release Date"]=f"20{random.randint(20, 23)}-{"0"*(2-len(str(mes)))+str(mes)}-{"0"*(2-len(str(dia)))+str(dia)}T00:00:00.000Z"
Combinacion.isnull().sum()
```

```python
Vistas=Combinacion["Hours Viewed"]
promedio=Vistas.mean()
for clave, valor in Vistas.items():
    if pd.isna(valor):
        Combinacion.at[clave, "Hours Viewed"]=promedio
Combinacion.isnull().sum()
```

```python
Valoraciones=Combinacion["Number of Ratings"]
for clave, valor in Valoraciones.items():
    if pd.isna(valor):
        Combinacion.at[clave, "Number of Ratings"]=float(random.randint(5, 100))
Combinacion.isnull().sum()
```

```python
Genero=Combinacion["Genre"]
Valores_no_nulos=Genero.dropna().to_list()

for clave, valor in Genero.items():
    if pd.isna(valor):
        indice=random.randint(0, len(Valores_no_nulos)-1)
        Combinacion.at[clave, "Genre"]=Valores_no_nulos[indice]
Combinacion.isnull().sum()
```

```python
Rateo=Combinacion["Rating"]
promedio=Rateo.mean()
for clave, valor in Rateo.items():
    if pd.isna(valor):
        Combinacion.at[clave, "Rating"]=promedio
Combinacion.isnull().sum()
```

```python
Palabras=Combinacion["Key Words"]
No_nulos=Palabras.dropna().to_list()
for clave, valor in Palabras.items():
    if pd.isna(valor):
        indice=random.randint(0, len(No_nulos)-1)
        Combinacion.at[clave, "Key Words"]=No_nulos[indice]
Combinacion.isnull().sum()
```

```python
Desc=Combinacion["Description"]
No_nulos=Desc.dropna().to_list()
for clave, valor in Desc.items():
    if pd.isna(valor):
        indice=random.randint(0, len(No_nulos)-1)
        Combinacion.at[clave, "Description"]=No_nulos[indice]
Combinacion.isnull().sum()
```

_Etapa 4 - Presentación de resultados_

```python
A=Combinacion[Combinacion['Available Globally?']=="Yes"]
B=Combinacion["Available Globally?"]
Cf=len(A)
Ct=len(B)
Porcentaje=(Cf/Ct)*100
with open(archivos+"/a.txt", "w", encoding="utf-8") as archivo1:
    archivo1.write(f"El porcentaje de películas/series que están disponibles a nivel es {round(Porcentaje, 4)}%\n")
```

```python
# Convertir la columna "Release Date" a objetos datetime en pandas (para el ordenamiento por fechas)
# Con "format='mixed'" se inferira el formato de la fecha para cada uno de los valores de la columna en el DataFrame
# Asigna la zona horaria UTC a las fechas convertidas (se aplico esto para prevenir un posible error)
Combinacion['Release date']=pd.to_datetime(Combinacion["Release Date"], format="mixed", utc=True)
# Ordenar las fechas
Ordenamiento=Combinacion.sort_values(by='Release date')
```

```python
with open(archivos + "/b.txt", "wb") as archivo2:
    # Guardar las 10 películas más antiguas en el archivo
    archivo2.write("\t\t\t10 Peliculas antiguas\n".encode('utf-8'))
    archivo2.write(Ordenamiento["Title"].head(10).to_string().encode('utf-8') + b"\n")

    # Guardar las 10 películas más recientes en el archivo
    archivo2.write("\t\t10 Peliculas recientes\n".encode('utf-8'))
    archivo2.write(Ordenamiento["Title"].tail(10).to_string().encode('utf-8') + b"\n")
```
